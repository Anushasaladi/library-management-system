```python
class Library() :
    def __init__(self):
        self.books = []
        for i in range(10):
            row = []
            for j in range(10):
                row.append(None)
            self.books.append(row) 
        self.deleted = []
    def insert(self):
        n = int(input("Enter number of books you want to insert: "))
        for i in range(n):
            book = {
                "Title": input("Enter the book name: "),
                "Author": input("Enter the author name: "),
                "Price": int(input("Enter the price: ")),
                "Genre": input("Enter the genre: "),
                "Copies": 1
            }
            found_duplicate = False  # inside for i loop!
            for k in range(10):
                for l in range(10):
                    if (self.books[k][l] is not None and self.books[k][l]["Title"].lower() == book["Title"].lower() and self.books[k][l]["Author"].lower() == book["Author"].lower()):
                        self.books[k][l]["Copies"] += 1
                        found_duplicate = True
                        break
                if found_duplicate:
                    break
            if not found_duplicate:  # inside for i loop!
                if self.deleted:
                    row, col = self.deleted.pop()
                    self.books[row][col] = book
                else:
                    found_slot = False
                    for r in range(10):
                        for c in range(10):
                            if self.books[r][c] is None:
                                self.books[r][c] = book
                                found_slot = True
                                break
                        if found_slot:
                            break
                    if not found_slot:
                        print("Library is full! Cannot insert more books.")
    def delete(self, title, author):
        for i in range(10):
            for j in range(10):
                if (self.books[i][j] is not None and
                    self.books[i][j]["Title"].lower() == title.lower() and
                    self.books[i][j]["Author"].lower() == author.lower()):
                
                    if self.books[i][j]["Copies"] > 1:
                        self.books[i][j]["Copies"] -= 1
                        print("One copy deleted. Remaining:", self.books[i][j]["Copies"])
                    else:
                        self.books[i][j] = None
                        self.deleted.append((i, j))
                        print("Book completely removed from library")
                    return  

        print(title, "book not available")
    def search(self,title):
        found =False
        for i in range(10):
            for j in range(10):
                if self.books[i][j] is not None and self.books[i][j]["Title"].lower() == title.lower():
                    found = True
                    print("Title:",self.books[i][j]["Title"])
                    print("Author:",self.books[i][j]["Author"])
                    print("Price:",self.books[i][j]["Price"])
                    print("Genre:",self.books[i][j]["Genre"])
                    print("No of Copies:",self.books[i][j]["Copies"])
                    print("Location: Row", i, "Column", j)
                    print("-" * 40)
        if not found:
            print(title,"book not found")
    def get_statistics(self):
        sum1 = 0
        sums = 0
        author_count = {}

        for i in range(10):
            for j in range(10):
                if self.books[i][j] is not None:
                    sum1 += 1
                    sums += self.books[i][j]["Copies"]

                    author = self.books[i][j]["Author"].lower()

                    if author in author_count:
                        author_count[author] += self.books[i][j]["Copies"]
                    else:
                        author_count[author] = self.books[i][j]["Copies"]

        print("Unique books in library are:", sum1)
        print("No of total books:", sums)

        for author, count in author_count.items():
            print(author, ":", count, "books")
    def display(self):
        found = False
        for i in range(10):
            for j in range(10):
                if self.books[i][j] is not None:
                    found = True
                    print("Title:", self.books[i][j]["Title"])
                    print("Author:", self.books[i][j]["Author"])
                    print("Price:", self.books[i][j]["Price"])
                    print("Genre:", self.books[i][j]["Genre"])
                    print("Copies:", self.books[i][j]["Copies"])
                    print("-" * 40)
        if not found:
            print("Library is empty")
lib = Library()
while True:
    print("\n===== Library Management System =====")
    print("1. Insert Books")
    print("2. Delete Book")
    print("3. Search Book")
    print("4. Display All Books")
    print("5. Statistics")
    print("6. Exit")
    choice = input("Enter your choice: ")

    match choice:
        case "1":
            lib.insert()
        case "2":
            title = input("Enter book title to delete: ")
            author = input("Enter author name: ")
            lib.delete(title,author)
        case "3":
            title = input("Enter book title to search: ")
            lib.search(title)
        case "4":
            lib.display()
        case "5":
            lib.get_statistics()
        case "6":
            print("Goodbye!")
            break
        case _:
            print("Invalid choice! Please enter 1-6")
            
```
